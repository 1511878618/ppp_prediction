```python
import json
```

```python
ukbData = "MulitOmicsDisease/"

from pathlib import Path

json_save_folder = (
    "/home/xutingfeng/ukb/project/ppp_prediction/MulitOmicsDisease/jsonData"
)
Path(json_save_folder).mkdir(parents=True, exist_ok=True)
```

```python
n_bootstrap = 500

# defined in the json
for disease_file_name in [
    "T2D_Coding_Amit_NG2018.feather",
    "CAD_v3.feather",
    "cad_xtf_always.feather",
]:

    ProtTrainDir = f"{ukbData}/traindata/Prot.feather"
    MetaTrainDir = f"{ukbData}/traindata/Meta.feather"
    RFTrainDir = f"{ukbData}/traindata/RF.feather"
    heldOutDataDir = f"{ukbData}/traindata/heldout.feather"

    phenoDataDir = f"{ukbData}/omicsData/phenos.feather"
    diseaseDataDir = f"{ukbData}/disease/{disease_file_name}"

    cov = ["age", "sex"]

    params_json = {
        "omicsData": {
            "Prot": {
                "name": "Prot",
                "path": ProtTrainDir,
                "feature": None,
            },
            "Meta": {
                "name": "Meta",
                "path": MetaTrainDir,
                "feature": None,
            },
            "AgeSex": {
                "name": "AgeSex",
                "path": RFTrainDir,
                "feature": None,  # None is ok
            },
        },
        "heldOutData": {
            "name": "heldOut",
            "path": heldOutDataDir,
        },
        "diseaseData": {
            "path": diseaseDataDir,
            "label": "event",
            "date": "date",
        },
        "phenosData": {
            "name": "phenos",
            "path": phenoDataDir,
        },
        "modelConfig": {
            "Prot": {
                "name": "Prot",  # name of the model for save
                "model": ["lasso"],  # not work now
                "feature": None,  # feature to use, None is all
                "cov": cov,  # covariate to use
                "cv": 10,  # cv
                "n_bootstrap": n_bootstrap,
            },
            "Meta": {
                "name": "Meta",
                "model": ["lasso"],
                "feature": None,
                "cov": cov,
                "cv": 10,
                "n_bootstrap": n_bootstrap,
            },
            "AgeSex": {
                "name": "AgeSex",
                "model": ["lasso"],
                "feature": ["age", "sex"],
                "cov": None,
                "cv": 10,
                "n_bootstrap": None,
            },
        },
    }
    json.dump(
        params_json, open(f"{json_save_folder}/{disease_file_name}.json", "w"), indent=4
    )
```
